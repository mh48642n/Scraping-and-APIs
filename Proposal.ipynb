## Okun's Law: Intricacies of the Manufacturing and Production Industry  

### Introduction

One of the most promiment macroeconomic theories is Okun's Law, that depicts the relationship between the change in output and the change in employment. This idea can be reflected on the macroeconomic scale by comparing the percent changes for both employment and gross domestic product prespectively. Of the literature that investigates the theory of Okun there has been confirmation that the theory between employment and output holds up across multiple countries(Obst, 2021)(Dixon et al, 2017). However, I am concerned on how this idea holds up within a single industry like manufacturing and production. As more human tasks have been shifted to technology and domestic business conditions reduce the likelihood of production and manufacturing domestically it is important to determine the significance of changes of employment on output for this industry.

to each other with both level and differenced data. From there I will use granger casaulity
